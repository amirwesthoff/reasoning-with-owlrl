```python
import rdflib
from owlrl import DeductiveClosure, OWLRL_Semantics

# Create a Graph
g = rdflib.Graph()

# Define the ontology in Turtle format
ontology_turtle = """
@prefix : <http://example.org/> .
@prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#> .
@prefix owl: <http://www.w3.org/2002/07/owl#> .
@prefix rdfs: <http://www.w3.org/2000/01/rdf-schema#> .

:Person a owl:Class .
:Man a owl:Class ;
    rdfs:subClassOf :Person .
:Woman a owl:Class ;
    rdfs:subClassOf :Person .

:John a :Man .
:Jane a :Woman .
"""

# Parse the ontology
g.parse(data=ontology_turtle, format="turtle")

# Print initial triples
print("Initial Triples:")
for s, p, o in g:
    print(f"{s} {p} {o}")

# Manually create a copy of the graph before reasoning
g_before_reasoning = rdflib.Graph()
for triple in g:
    g_before_reasoning.add(triple)

# Perform OWL RL reasoning on the original graph
DeductiveClosure(OWLRL_Semantics).expand(g)

# Calculate the inferred triples
inferred_triples = set(g) - set(g_before_reasoning)

# Print inferred triples
print("\nInferred Triples:")
for s, p, o in inferred_triples:
    print(f"{s} {p} {o}")
```
